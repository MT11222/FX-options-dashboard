```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create the data
data = {
    'Partners': ['1 partner', '2-4 partners', '5-9 partners', '10-14 partners', '15 or more partners'],
    'Women 25-49': [17.7, 29.2, 28.6, 11.6, 12.9],
    'Men 25-49': [11.2, 22.3, 25.8, 12.5, 28.3]
}

# Create DataFrame
df = pd.DataFrame(data)

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot both lines
ax.plot(df['Partners'], df['Women 25-49'], marker='o', linewidth=2, label='Women 25-49')
ax.plot(df['Partners'], df['Men 25-49'], marker='s', linewidth=2, label='Men 25-49')

# Add labels and title
ax.set_xlabel('Number of Partners')
ax.set_ylabel('Percent')
ax.set_title('Percentage of Women and Men Aged 25-49 by Number of Partners (2015-2019)')

# Add grid for better readability
ax.grid(True, linestyle='--', alpha=0.7)

# Add legend
ax.legend()

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create the original data
data_original = {
    'Partners': ['1 partner', '2-4 partners', '5-9 partners', '10-14 partners', '15 or more partners'],
    'Women 25-49': [17.7, 29.2, 28.6, 11.6, 12.9],
    'Men 25-49': [11.2, 22.3, 25.8, 12.5, 28.3]
}

# Create DataFrame for original data
df_original = pd.DataFrame(data_original)

# Create new data with combined categories
# For the "10 or more partners" category, we add the percentages of "10-14 partners" and "15 or more partners"
data_combined = {
    'Partners': ['1 partner', '2-4 partners', '5-9 partners', '10 or more partners'],
    'Women 25-49': [17.7, 29.2, 28.6, 11.6 + 12.9],  # Adding 10-14 and 15+ for women
    'Men 25-49': [11.2, 22.3, 25.8, 12.5 + 28.3]     # Adding 10-14 and 15+ for men
}

# Create DataFrame for combined data
df_combined = pd.DataFrame(data_combined)

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot both lines with the combined data
ax.plot(df_combined['Partners'], df_combined['Women 25-49'], marker='o', linewidth=2, label='Women 25-49')
ax.plot(df_combined['Partners'], df_combined['Men 25-49'], marker='s', linewidth=2, label='Men 25-49')

# Add labels and title
ax.set_xlabel('Number of Partners')
ax.set_ylabel('Percent')
ax.set_title('Percentage of Women and Men Aged 25-49 by Number of Partners (2015-2019)')

# Add grid for better readability
ax.grid(True, linestyle='--', alpha=0.7)

# Add legend
ax.legend()

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
